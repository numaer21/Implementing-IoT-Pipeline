## Environment setup

We begin by importing the required libraries. TensorFlow is used for building neural networks, NumPy for numerical operations, Matplotlib for visualization, and scikit-learn for dataset generation and splitting.

```

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

The imports above must be executed before running any other cells, as later code depends on these libraries.

## Tensors and basic numerical structures

This section demonstrates how TensorFlow represents scalars, vectors, matrices, and higher-rank tensors. Understanding shapes and ranks is essential for debugging deep learning models.

```

scalar = tf.constant(7)
vector = tf.constant([10, 20, 30])
matrix = tf.constant([[1., 2.], [3., 4.]])
tensor = tf.constant([[[1],[2]], [[3],[4]]])

print("Shapes:", scalar.shape, vector.shape, matrix.shape, tensor.shape)
```

The printed shapes confirm the rank of each tensor. Shape mismatches are one of the most common sources of errors in neural network development.

## Regression with a neural network

We train a simple neural network to learn the linear relationship y = x + 10. This example illustrates the standard Keras workflow: create, compile, fit, and evaluate.

```

X = np.arange(-20, 21, 2, dtype=np.float32)
y = X + 10

X_train, y_train = X[:20], y[:20]
X_test, y_test = X[20:], y[20:]

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(loss="mae", optimizer="sgd")
model.fit(X_train, y_train, epochs=200, verbose=0)

y_pred = model.predict(X_test).squeeze()

plt.scatter(X_train, y_train, label="Train")
plt.scatter(X_test, y_test, label="Test")
plt.scatter(X_test, y_pred, label="Prediction")
plt.legend()
plt.title("Regression: Learning y = x + 10")
plt.show()
```

The plotted predictions closely follow the true test values, showing that the network has successfully learned and generalized the linear function.

## Binary classification on non-linear data

In this section, we solve a binary classification problem using a non-linear dataset. Hidden layers with ReLU activation allow the model to learn curved decision boundaries.

```

from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split

X, y = make_circles(n_samples=1000, noise=0.03, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(X_train, y_train, epochs=25, verbose=0)

loss, acc = model.evaluate(X_test, y_test, verbose=0)
print("Test accuracy:", acc)
```

The achieved accuracy demonstrates that non-linear activation functions enable neural networks to solve problems that linear models cannot.
